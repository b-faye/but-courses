#### Inheritance and overriding

Inheritance is a mechanism that allows you to create a hierarchy of classes that share a set of properties and methods by deriving a class from a parent class.

In the following, we consider  ```Person``` class which will be the parent class:  

````python
class Person():
  """
  name: Person's name
  """
  def __init__(self, name):
    self.name = name
    self.age = None

  def __str__(self):
    return("Person: {}".format(self.name))

  def set_age(self, age):
    self.age = age

  def get_age(self):
    return(self.age)
````

````python
person = Person(name="Robert")
print(person)
````

````python
person.set_age(42)
person.get_age()
````

We define now a child class ```Student``` which inherits from ```Person``` class with some extra properties:

````python
class Student(Person):
  """
  name: Person's name
  age: Person's age
  """
  def __init__(self, name, field):
    self.field = field
    self.year = None
    super().__init__(name)

  def __str__(self):
    return("Student {}, field {}".format(self.name, self.field))

  def set_year(self, year):
    self.year = year

  def get_year(self):
    return(self.year)
````

This child class ```Student``` inherits the attributes ```name``` and ```age``` and the methods ```set_age()``` and ```get_age()``` from the parent class ``` Person```:

````python
student = Student(name="Maëlle", field="Computer science")
print(student)
````

````python
student.set_age(22)
student.get_age()
````

But it has two extra attributes ```field``` and ```year``` and two extra methods ```set_year()``` and ```get_year()```:

````python
student.set_year(4)
student.get_year()
````


  The method ```__str__``` that it prints the name and the field overrides inheritance. The ```super()``` function is used to refer to the parent class or superclass. It allows to call methods defined in the superclass from the subclass, enabling you to extend and customize the functionality inherited from the parent class.

1. Define a function ```fibonnacci(n)``` that returns the list of $n\geq 0$ first elements of fibonnacci sequence defined as:
$$
 \begin{cases}
u_0 = 0,\\
u_1 = 1, \\
u_{n+2} = u_{n+1} + u_n.
\end{cases}
$$

````python
# Answer
def fibonacci(n):
    if n < 0:
        raise ValueError("n should be positive")
    elif n >= 0 and n<2:
      return list(range(0,n+1))
    else:
      # Initialize the Fibonacci sequence with the first two elements
      sequence = [0, 1]

      # Generate the Fibonacci sequence up to the n-th element
      for i in range(2, n):
          sequence.append(sequence[i - 1] + sequence[i - 2])
      return sequence

# Example usage:
if __name__ == "__main__":
    n = 10
    fib_sequence = fibonacci(n)
    print(f"Fibonacci Sequence of the first {n} elements:")
    print(fib_sequence)
````

2. Define a function ```pascal(n)``` that returns a list that contains elements of the $n$-th line of the [Pascal's triangle](https://en.wikipedia.org/wiki/Pascal%27s_triangle).   

````python
# Answer
def pascal(n):
    if n < 0:
        raise ValueError("n should be a non-negative integer")

    # Initialize the first row of Pascal's triangle
    row = [1]

    # Generate the nth row of Pascal's triangle
    for i in range(1, n + 1):
        next_element = row[-1] * (n - i + 1) // i
        row.append(next_element)
    return row

# Example usage:
if __name__ == "__main__":
    n = 2  # Change n to the desired line number
    pascal_row = pascal(n)
    print(f"Pascal's Triangle - Line {n}:")
    print(pascal_row)
````

3. Let $(u_n)_{n\in\mathbb{N}}$ and $(v_n)_{n\in\mathbb{N}}$ be two numerical sequences defined as $u_0=1$, $v_0=1$, and $∀n\geq 0$:
$$
u_{n+1} = u_n + v_n,\quad v_{n+1}=2u_n - v_n,
$$
Compute $u_{100}$ and $v_{100}$

````python
# Answer
def compute_u_v(n):
    u = [1]  # Initialize u_0
    v = [1]  # Initialize v_0

    for i in range(1, n + 1):
        u_n = u[i - 1] + v[i - 1]
        v_n = 2 * u[i - 1] - v[i - 1]
        u.append(u_n)
        v.append(v_n)
    return u, v

# Example usage:
if __name__ == "__main__":
    n = 100  # Change n to the desired value
    u_sequence, v_sequence = compute_u_v(n)

    print(f"u_{n}:", u_sequence[-1])
    print(f"v_{n}:", v_sequence[-1])
````

4. Let $n_0,K \geq 1$ be two integers. Let $v=(v_k)_{k\in\mathbb{N}}$ be the sequence defined by $v_0=n_0$ and

$$
v_{k+1} =  \begin{cases}
			3v_k +1 & \text{if $v_k$ is odd,}\\
            \frac{v_k}{2} & \text{if $v_k$ is even,}
		 \end{cases}
$$
Define a function ```vk(n0, K)``` that returns $K$ first elements of the sequence $v$.

````python
# Answer
def vk(n0, K):
    if K < 1:
        raise ValueError("K should be a positive integer")

    sequence = [n0]  # Initialize with n0

    while len(sequence) < K:
        if sequence[-1] % 2 == 0:
            # If the last element is even, divide by 2
            next_element = sequence[-1] // 2
        else:
            # If the last element is odd, apply the 3v + 1 rule
            next_element = 3 * sequence[-1] + 1
        sequence.append(next_element)

    return sequence

# Example usage:
if __name__ == "__main__":
    n0 = 5  # Change n0 to the desired starting value
    K = 10  # Change K to the desired number of elements
    v_sequence = vk(n0, K)

    print(f"v_{K} sequence:")
    print(v_sequence)
````

For $K = 1000$ and for every $n_0\in\{10,100,1000,10000\}$ print the last five values.

````python
# Answer
if __name__ == "__main__":
    K = 1000
    initial_values = [10, 100, 1000, 10000]

    for n0 in initial_values:
        v_sequence = vk(n0, K)

        print(f"n0 = {n0}, Last Five Values of v_{K} sequence:")
        print(v_sequence[-5:])
````

5. Define a function ```wallis(n)``` that computes an approximation of $\pi$ using the [Wallis product](https://en.wikipedia.org/wiki/Wallis_product).

````python
# Answer
def wallis(n):
    if n < 1:
        raise ValueError("n should be a positive integer")

    pi_estimate = 1.0

    for i in range(1, n + 1):
        term = (4 * i ** 2) / ((4 * i ** 2) - 1)
        pi_estimate *= term

    # Multiply by 2 to complete the product
    pi_estimate *= 2

    return pi_estimate

# Example usage:
if __name__ == "__main__":
    n = 10000  # Change n to the desired number of iterations
    pi_approximation = wallis(n)

    print(f"Approximation of Pi using Wallis Product with {n} iterations:")
    print(pi_approximation)
````

We recall, that the Body Mass Index (BMI) given the Weight (kg) and Heigth (m) is computed as follows:
$$
BMI = \frac{Weight}{Height^2}
$$
1. Define a function ```body_mass_index(weight,height)``` that returns a tuple ```(bmi, interpretation)```  that contains the BMI and its interpretation (i.e., Underweigth, Nomal, or Obese). Please refer to [https://en.wikipedia.org/wiki/Body_mass_index](https://en.wikipedia.org/wiki/Body_mass_index) to interpret the BMI.  

````python
# Answer
def body_mass_index(weight, height):
    if weight <= 0 or height <= 0:
        raise ValueError("Weight and height must be positive values")

    bmi = weight / (height ** 2)

    interpretation = ""
    if bmi < 18.5:
        interpretation = "Underweight"
    elif 18.5 <= bmi < 24.9:
        interpretation = "Normal Weight"
    else:
        interpretation = "Obese"

    return bmi, interpretation

# Example usage:
if __name__ == "__main__":
    weight = 70  # Change weight to the desired weight in kg
    height = 1.75  # Change height to the desired height in meters

    bmi, interpretation = body_mass_index(weight, height)

    print(f"BMI: {bmi}")
    print(f"Interpretation: {interpretation}")
````

2. Write a script that adds the fields ```Bmi``` and ```Indication``` which are the BMI and its interpretation to each dictionnary defined above.

````python
# Answer
# Define the list of dictionaries of persons
Persons = [{"Name": "Robert", "Weight": 95, "Height": 1.81},
           {"Name": "Alice", "Weight": 48, "Height": 1.69},
           {"Name": "Maëlle", "Weight": 61, "Height": 1.57},
           {"Name": "Maxime", "Weight": 140, "Height": 1.98}]

# Define the body_mass_index function
def body_mass_index(weight, height):
    if weight <= 0 or height <= 0:
        raise ValueError("Weight and height must be positive values")

    bmi = weight / (height ** 2)

    interpretation = ""
    if bmi < 18.5:
        interpretation = "Underweight"
    elif 18.5 <= bmi < 24.9:
        interpretation = "Normal Weight"
    else:
        interpretation = "Obese"

    return bmi, interpretation

# Add "Bmi" and "Indication" fields to each dictionary in Persons
for person in Persons:
    bmi, indication = body_mass_index(person["Weight"], person["Height"])
    person["Bmi"] = bmi
    person["Indication"] = indication

# Print the updated list of persons
for person in Persons:
    print("Name:", person["Name"])
    print("Weight:", person["Weight"], "kg")
    print("Height:", person["Height"], "m")
    print("BMI:", person["Bmi"])
    print("Indication:", person["Indication"])
    print()
````

### Exercice 2: Power iteration method

The objective of this excercice is to compute the eigenvector of a matrix associated to its largest eigenvector using the power iteration method: Given a squared matrix $A$ of shape $n\times n$ and a vector $x_0\in\mathbb{R}^n$, then the sequence of vectors $(x_k)_{k\in\mathbb{N}}$ defined as:
$$
x_{k+1}=\frac{Ax_k}{\Vert A x_k\Vert},
$$
converges to the largest eigenvector in absolute value (we admit this result).
  

1. Define a function ```power_iteration(A, x0, k)``` that returns $x_k$.



````python
# Answer

import numpy as np

def power_iteration(A, x0, k):
    x = x0
    for _ in range(k):
        # Compute the matrix-vector multiplication Ax
        Ax = np.dot(A, x)

        # Compute the norm of Ax
        norm_Ax = np.linalg.norm(Ax)

        # Update the eigenvector approximation
        x = Ax / norm_Ax

    return x

# Example usage
n = 5  # Size of the matrix
A = np.random.randn(n, n)  # Replace with your matrix A
x0 = np.random.randn(n)     # Initial vector
k = 100                     # Number of iterations

# Call the power_iteration function
dominant_eigenvector = power_iteration(A, x0, k)
print("Dominant eigenvector:\n", dominant_eigenvector)
````

4. Write a script that allows to compute the largest eigenvalue with its associated eigenvector of a matrix given precision value.

````python
# Answer
import numpy as np

def power_iteration(A, precision):
    n = A.shape[0]  # Dimension of the matrix
    x = np.random.randn(n)  # Initial vector

    while True:
        # Compute the matrix-vector multiplication Ax
        Ax = np.dot(A, x)

        # Compute the norm of Ax
        norm_Ax = np.linalg.norm(Ax)

        # Update the eigenvector approximation
        new_x = Ax / norm_Ax

        # Check for convergence
        if np.linalg.norm(new_x - x) < precision:
            break

        x = new_x

    # Compute the eigenvalue associated with the eigenvector
    eigenvalue = np.dot(np.dot(A, x), x) / np.dot(x, x)

    return eigenvalue, x

# Given matrix A
A = np.array([[0.5, 0.5],
              [0.2, 0.8]])

# Desired precision for convergence
precision = 1e-6

# Compute the largest eigenvalue and associated eigenvector using power_iteration
largest_eigenvalue, largest_eigenvector = power_iteration(A, precision)

print("Largest eigenvalue:", largest_eigenvalue)
print("Associated eigenvector:", largest_eigenvector)
````

````python
# Answer
import numpy as np

class SquaredMatrix(np.matrix):
    def __new__(cls, values):
        if values.shape[0] != values.shape[1]:
            raise ValueError("Input matrix must be square")
        return super().__new__(cls, values)

    def rank(self):
        # Calculate and return the rank of the matrix
        return np.linalg.matrix_rank(self)

    def is_invertible(self):
        # Check if the matrix is invertible (non-singular)
        return self.rank() == self.shape[0]

    def inverse(self):
        # Check if the matrix is invertible, and if so, return its inverse
        if self.is_invertible():
            return np.linalg.inv(self)
        else:
            raise ValueError("Matrix is not invertible")

    def eigenvalues(self):
        # Calculate and return the eigenvalues of the matrix
        return np.linalg.eigvals(self)

    def eigenvectors(self):
        # Calculate and return the eigenvalues and eigenvectors of the matrix
        eigenvalues, eigenvectors = np.linalg.eig(self)
        return eigenvalues, eigenvectors

# Example usage:
if __name__ == "__main__":
    # Create a squared matrix
    matrix_data = np.array([[1, 2], [3, 4]])
    squared_matrix = SquaredMatrix(matrix_data)

    print("Squared Matrix:")
    print(squared_matrix)

    print("Rank:", squared_matrix.rank())
    print("Is Invertible:", squared_matrix.is_invertible())

    try:
        inverse_matrix = squared_matrix.inverse()
        print("Inverse Matrix:")
        print(inverse_matrix)
    except ValueError as e:
        print("Error:", e)

    eigenvalues, eigenvectors = squared_matrix.eigenvectors()
    print("Eigenvalues:")
    print(eigenvalues)
    print("Eigenvectors:")
    print(eigenvectors)
````

&||A||_1 \,\, = \max_{1\leq j\leq n} \sum_{i=1}^{n}|a_{i,j}|, \\
&||A||_2 \,\, = \sqrt{\rho(A^tA)}, \\
&||A||_\infty = \max_{1\leq i\leq n} \sum_{j=1}^{n}|a_{i,j}|,
\end{align*}
$$
where $\rho(\cdot)$ design the spectral radius, i.e., the largest (in absolute value) eigenvalue of a matrix. Define the method ```norm_f(ind)``` that returns the norm given the indice ```ind``` (the indice is a string with possible values: ```"F"```,```"1"```,```"2"```, and ```"inf"```).


````python
# Answer
import numpy as np

class SquaredMatrix(np.matrix):
    def __new__(cls, values):
        if values.shape[0] != values.shape[1]:
            raise ValueError("Input matrix must be square")
        return super().__new__(cls, values)

    def norm_f(self, ind):
        if ind == "F":
            # Frobenius norm
            return np.linalg.norm(self, ord='fro')
        elif ind == "1":
            # 1-norm
            return np.linalg.norm(self, ord=1)
        elif ind == "2":
            # 2-norm (spectral norm)
            return np.linalg.norm(self, ord=2)
        elif ind == "inf":
            # Infinity norm
            return np.linalg.norm(self, ord=np.inf)
        else:
            raise ValueError("Invalid norm index. Use 'F', '1', '2', or 'inf'.")

# Example usage:
if __name__ == "__main__":
    # Create a squared matrix
    matrix_data = np.array([[1, 2], [3, 4]])
    squared_matrix = SquaredMatrix(matrix_data)

    print("Squared Matrix:")
    print(squared_matrix)

    # Calculate different norms
    print("Frobenius Norm (||A||_F):", squared_matrix.norm_f("F"))
    print("1-Norm (||A||_1):", squared_matrix.norm_f("1"))
    print("2-Norm (||A||_2):", squared_matrix.norm_f("2"))
    print("Infinity Norm (||A||_inf):", squared_matrix.norm_f("inf"))
````

### Exercise 4

Comparing Two Datasets

Select two datasets of your choice (e.g., salary data for different professions) or generate two synthetic datasets. Perform the following analyses:

1. Load and inspect the datasets.<br>
2. Calculate basic summary statistics for each dataset.<br>
3. Create histograms or kernel density plots to compare the distributions of a shared numerical variable.<br>
4. Calculate correlation coefficients if applicable.<br>
Analyze and discuss the similarities and differences between the datasets.


````python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate two synthetic datasets
np.random.seed(42)
data1 = np.random.normal(10, 2, 100)
data2 = np.random.normal(12, 3, 100)

# Calculate summary statistics for each dataset
summary_stats1 = pd.Series(data1).describe()
summary_stats2 = pd.Series(data2).describe()

# Create kernel density plots to compare distributions
plt.figure(figsize=(12, 6))
sns.kdeplot(data1, label='Dataset 1')
sns.kdeplot(data2, label='Dataset 2')
plt.title("Comparison of Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()

# Calculate correlation coefficient between datasets
correlation = np.corrcoef(data1, data2)[0, 1]
print("Correlation Coefficient:", correlation)
````

## Exercise 5

Simulating Random Events

Simulate a real-world scenario involving random events. For example, you can simulate the number of customers arriving at a store each hour. Perform the following analyses:

1. Define the simulation parameters (e.g., arrival rate).<br>
2. Simulate the random events for a certain time period.<br>
3. Calculate descriptive statistics for the simulated data, such as mean and variance.<br>
3. Create histograms or line plots to visualize the simulated events over time.<br>
4. Discuss insights gained from the simulation and how they relate to the real-world scenario.


````python
import numpy as np
import matplotlib.pyplot as plt

# Simulation parameters
hours = 24
arrival_rate = 5  # Average number of arrivals per hour

# Simulate random events (number of arrivals per hour)
np.random.seed(42)
arrivals = np.random.poisson(arrival_rate, hours)

# Calculate summary statistics
mean_arrivals = np.mean(arrivals)
var_arrivals = np.var(arrivals)

# Create a line plot to visualize simulated events over time
plt.plot(range(hours), arrivals, marker='o')
plt.title("Simulated Arrivals Over Time")
plt.xlabel("Hour")
plt.ylabel("Number of Arrivals")
plt.xticks(range(hours))
plt.grid(True)
plt.show()

print("Mean Arrivals:", mean_arrivals)
print("Variance of Arrivals:", var_arrivals)
````
