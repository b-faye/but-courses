1. Define a function ```fibonnacci(n)``` that returns the list of $n\geq 2$

1.   List item
2.   List item

first elements of fibonnacci sequence defined as:
$$
 \begin{cases}
u_0 = 0,\\
u_1 = 1, \\
u_{n+2} = u_{n+1} + u_n.
\end{cases}
$$

````python
# Answer
def fibonacci(n):
    if n < 2:
        raise ValueError("n should be greater than or equal to 2")

    # Initialize the Fibonacci sequence with the first two elements
    sequence = [0, 1]

    # Generate the Fibonacci sequence up to the nth element
    for i in range(2, n):
        next_element = sequence[i - 1] + sequence[i - 2]
        sequence.append(next_element)

    return sequence

# Example usage:
if __name__ == "__main__":
    n = 10  # Change n to the desired number of elements
    fib_sequence = fibonacci(n)
    print(f"Fibonacci Sequence of the first {n} elements:")
    print(fib_sequence)
````

2. Define a function ```pascal(n)``` that returns a list that contains elements of the $n$-th line of the [Pascal's triangle](https://en.wikipedia.org/wiki/Pascal%27s_triangle).   

````python
# Answer
def pascal(n):
    if n < 0:
        raise ValueError("n should be a non-negative integer")

    # Initialize the first row of Pascal's triangle
    row = [1]

    # Generate the nth row of Pascal's triangle
    for i in range(1, n + 1):
        next_element = row[-1] * (n - i + 1) // i
        row.append(next_element)

    return row

# Example usage:
if __name__ == "__main__":
    n = 5  # Change n to the desired line number
    pascal_row = pascal(n)
    print(f"Pascal's Triangle - Line {n}:")
    print(pascal_row)
````

3. Let $(u_n)_{n\in\mathbb{N}}$ and $(v_n)_{n\in\mathbb{N}}$ be two numerical sequences defined as $u_0=1$, $v_0=1$, and $∀n\geq 0$:
$$
u_{n+1} = u_n + v_n,\quad v_{n+1}=2u_n - v_n,
$$
Compute $u_{100}$ and $v_{100}$

````python
# Answer
def compute_u_v(n):
    u = [1]  # Initialize u_0
    v = [1]  # Initialize v_0

    for i in range(1, n + 1):
        u_n = u[i - 1] + v[i - 1]
        v_n = 2 * u[i - 1] - v[i - 1]
        u.append(u_n)
        v.append(v_n)

    return u, v

# Example usage:
if __name__ == "__main__":
    n = 100  # Change n to the desired value
    u_sequence, v_sequence = compute_u_v(n)

    print(f"u_{n}:", u_sequence[-1])
    print(f"v_{n}:", v_sequence[-1])
````

4. Let $n_0,K \geq 1$ be two integers. Let $v=(v_k)_{k\in\mathbb{N}}$ be the sequence defined by $v_0=n_0$ and

$$
v_{k+1} =  \begin{cases}
			3v_k +1 & \text{if $v_k$ is odd,}\\
            \frac{v_k}{2} & \text{if $v_k$ is even,}
		 \end{cases}
$$
Define a function ```vk(n0, K)``` that returns $K$ first elements of the sequence $v$.

````python
# Answer
def vk(n0, K):
    if K < 1:
        raise ValueError("K should be a positive integer")

    sequence = [n0]  # Initialize with n0

    while len(sequence) < K:
        if sequence[-1] % 2 == 0:
            # If the last element is even, divide by 2
            next_element = sequence[-1] // 2
        else:
            # If the last element is odd, apply the 3v + 1 rule
            next_element = 3 * sequence[-1] + 1
        sequence.append(next_element)

    return sequence

# Example usage:
if __name__ == "__main__":
    n0 = 5  # Change n0 to the desired starting value
    K = 10  # Change K to the desired number of elements
    v_sequence = vk(n0, K)

    print(f"v_{K} sequence:")
    print(v_sequence)
````

For $K = 1000$ and for every $n_0\in\{10,100,1000,10000\}$ print the last five values.

````python
# Answer
if __name__ == "__main__":
    K = 1000
    initial_values = [10, 100, 1000, 10000]

    for n0 in initial_values:
        v_sequence = vk(n0, K)

        print(f"n0 = {n0}, Last Five Values of v_{K} sequence:")
        print(v_sequence[-5:])
````

5. Define a function ```wallis(n)``` that computes an approximation of $\pi$ using the [Wallis product](https://en.wikipedia.org/wiki/Wallis_product).

````python
# Answer
def wallis(n):
    if n < 1:
        raise ValueError("n should be a positive integer")

    pi_estimate = 1.0

    for i in range(1, n + 1):
        term = (4 * i ** 2) / ((4 * i ** 2) - 1)
        pi_estimate *= term

    # Multiply by 2 to complete the product
    pi_estimate *= 2

    return pi_estimate

# Example usage:
if __name__ == "__main__":
    n = 10000  # Change n to the desired number of iterations
    pi_approximation = wallis(n)

    print(f"Approximation of Pi using Wallis Product with {n} iterations:")
    print(pi_approximation)
````

We recall, that the Body Mass Index (BMI) given the Weight (kg) and Heigth (m) is computed as follows:
$$
BMI = \frac{Weight}{Height^2}
$$
1. Define a function ```body_mass_index(weight,height)``` that returns a tuple ```(bmi, interpretation)```  that contains the BMI and its interpretation (i.e., Underweigth, Nomal, or Obese). Please refer to [https://en.wikipedia.org/wiki/Body_mass_index](https://en.wikipedia.org/wiki/Body_mass_index) to interpret the BMI.  

````python
# Answer
def body_mass_index(weight, height):
    if weight <= 0 or height <= 0:
        raise ValueError("Weight and height must be positive values")

    bmi = weight / (height ** 2)

    interpretation = ""
    if bmi < 18.5:
        interpretation = "Underweight"
    elif 18.5 <= bmi < 24.9:
        interpretation = "Normal Weight"
    else:
        interpretation = "Obese"

    return bmi, interpretation

# Example usage:
if __name__ == "__main__":
    weight = 70  # Change weight to the desired weight in kg
    height = 1.75  # Change height to the desired height in meters

    bmi, interpretation = body_mass_index(weight, height)

    print(f"BMI: {bmi}")
    print(f"Interpretation: {interpretation}")
````

2. Write a script that adds the fields ```Bmi``` and ```Indication``` which are the BMI and its interpretation to each dictionnary defined above.

````python
# Answer
# Define the list of dictionaries of persons
Persons = [{"Name": "Robert", "Weight": 95, "Height": 1.81},
           {"Name": "Alice", "Weight": 48, "Height": 1.69},
           {"Name": "Maëlle", "Weight": 61, "Height": 1.57},
           {"Name": "Maxime", "Weight": 140, "Height": 1.98}]

# Define the body_mass_index function
def body_mass_index(weight, height):
    if weight <= 0 or height <= 0:
        raise ValueError("Weight and height must be positive values")

    bmi = weight / (height ** 2)

    interpretation = ""
    if bmi < 18.5:
        interpretation = "Underweight"
    elif 18.5 <= bmi < 24.9:
        interpretation = "Normal Weight"
    else:
        interpretation = "Obese"

    return bmi, interpretation

# Add "Bmi" and "Indication" fields to each dictionary in Persons
for person in Persons:
    bmi, indication = body_mass_index(person["Weight"], person["Height"])
    person["Bmi"] = bmi
    person["Indication"] = indication

# Print the updated list of persons
for person in Persons:
    print("Name:", person["Name"])
    print("Weight:", person["Weight"], "kg")
    print("Height:", person["Height"], "m")
    print("BMI:", person["Bmi"])
    print("Indication:", person["Indication"])
    print()
````

1. Plot in the same graph the functions $sin(kx)$ and $cos(kx)$ for $k=1,2,3$ and $x\in[0,2\pi]$. Add a title, set x and y axis to x and y respectively, add a legend in bottom left corner position and set the x axis labels to {$0,\,\frac{\pi}{2},\,\pi,\,\frac{3\pi}{1},\,2\pi$}.

````python
# Answer
import numpy as np
import matplotlib.pyplot as plt

# Define the values of k and x
k_values = [1, 2, 3]
x = np.linspace(0, 2 * np.pi, 1000)  # Generate x values from 0 to 2*pi

# Create a figure and axis
fig, ax = plt.subplots()

# Plot sin(kx) and cos(kx) for each k
for k in k_values:
    sin_kx = np.sin(k * x)
    cos_kx = np.cos(k * x)

    ax.plot(x, sin_kx, label=f'sin({k}x)')
    ax.plot(x, cos_kx, label=f'cos({k}x)')

# Set the title and axis labels
ax.set_title('Sine and Cosine Functions for k = 1, 2, 3')
ax.set_xlabel('x')
ax.set_ylabel('y')

# Set the x-axis ticks and labels
x_ticks = [0, np.pi / 2, np.pi, (3 * np.pi) / 2, 2 * np.pi]
x_labels = ['$0$', '$\\frac{\\pi}{2}$', '$\\pi$', '$\\frac{3\\pi}{2}$', '$2\\pi$']
ax.set_xticks(x_ticks)
ax.set_xticklabels(x_labels)

# Add a legend in the bottom left corner
ax.legend(loc='lower left')

# Display the plot
plt.show()
````

2. Define three methods ```get_diameter()```, ```get_perimeter()``` and ```get_area()``` that return  the diameter, the perimeter and the area of the circle.

````python
# Answer
import math

class Circle:
    def __init__(self, R):
        self.R = R

    def print_circle(self):
        print(f"This is a circle with radius {self.R}.")

    def get_diameter(self):
        diameter = 2 * self.R
        return diameter

    def get_perimeter(self):
        perimeter = 2 * math.pi * self.R
        return perimeter

    def get_area(self):
        area = math.pi * (self.R ** 2)
        return area

# Instantiate an object of the Circle class with R = 3
circle = Circle(3)

# Print the object
circle.print_circle()

# Get and print the diameter, perimeter, and area
print("Diameter:", circle.get_diameter())
print("Perimeter:", circle.get_perimeter())
print("Area:", circle.get_area())
````

3. Define a child class ```Sphere``` that inherits from the ```Circle``` class. Adapt the methods if necessary. What can you say about the methods.

````python
# Answer
import math

class Circle:
    def __init__(self, R):
        self.R = R

    def print_circle(self):
        print(f"This is a circle with radius {self.R}.")

    def get_diameter(self):
        diameter = 2 * self.R
        return diameter

    def get_perimeter(self):
        perimeter = 2 * math.pi * self.R
        return perimeter

    def get_area(self):
        area = math.pi * (self.R ** 2)
        return area

class Sphere(Circle):
    def __init__(self, R):
        super().__init__(R)

    def print_sphere(self):
        print(f"This is a sphere with radius {self.R}.")

    def get_diameter(self):
        diameter = 2 * self.R
        return diameter

    def get_surface_area(self):
        surface_area = 4 * math.pi * (self.R ** 2)
        return surface_area

    def get_volume(self):
        volume = (4/3) * math.pi * (self.R ** 3)
        return volume

# Instantiate an object of the Sphere class with R = 3
sphere = Sphere(3)

# Print the sphere object
sphere.print_sphere()

# Get and print the diameter, surface area, and volume of the sphere
print("Diameter:", sphere.get_diameter())
print("Surface Area:", sphere.get_surface_area())
print("Volume:", sphere.get_volume())
````
