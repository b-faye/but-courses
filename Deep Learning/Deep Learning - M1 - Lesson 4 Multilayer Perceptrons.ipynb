## **Artificial Neural Networks**

### **Introduction**

* Deep Learning is a branch of machine learning based on computational graph  models called ***Neural Networks***;

\\

* Neural Networks are **Mathematical Function** biologically inspired (invented by psychologist [Frank Rosenblatt](https://en.wikipedia.org/wiki/Frank_Rosenblatt) in 1960);

\\

* Biologically speaking, neurons are
  * cells able to make basic computation from input signals;
  * present in huge quantity (arround 100 billions in Human brain);
  * interconnected and communicate through electrical signals with different intensities;

* If one neuron **fires**, the signal is forwarded connected neurons otherwise the signal is stopped and the information is lost.
  
\\

![](https://drive.google.com/uc?export=view&id=1qzo261xXKNlB56avJmRgROt4u8qk3zIq)




### **Definition**

#### **Basic Architecture**

* Mathematically speaking, neurons are **scalars** grouped in layers forming **vectors**

\\



![](https://drive.google.com/uc?export=view&id=1b34bsaJV2FSF8UNBzD3vlKa0v9HRe0mL)


* Neural Networks are read from left to right:  
  
   * **First layer**: representation of intput $\bf{x}$;
    
   * **Middle layer**: also called **hidden layer** represents  **intermediary** transformation of $\bf{x}$;
    
   * **Last layer**: representation of output $\bf{y}$ (which can be scalar or vector).  
    

\\


* Connections between neurons are represented with arrows;

* Each arrow corresponds to scalar representing the **weight** of the source neuron over the target neuron;

* Weights relating between successive layers of neurons are **compiled** into **matrices**;  

\\

* Each neuron gets evaluated through three operations:

    1. weighted sum of entering neurons;

    2. to which we add a bias; 

    3. the result is wrapped with **non linear** function $a(x)$;


\\

* $a(x)$ is known as **activation function**:
  * Biologically, it emulates the mechanism of neuron firing;
  * Mathematically, it helps to learn non-linear function.


#### **Activation Functions**

* Activation functions represent crucial component in neural networks. Many have been proposed, logistic (also known as sigmoid) function is one of them

\\

![](https://drive.google.com/uc?export=view&id=1srYqF0mi9ztUzxA5h1HGonIdereq85v-)



#### **Neural Networks in a Nutshell**

* Neural Networks are mathematical functions mapping between two vector spaces;


* Mapping happens through series of **non-linear** transformations; 


* Stages of transformation are represented through **intermediary** layers/vectors;


* Passing from one layer to another consists in applying **activation function** over the affine transformation of the previous layer;

## **Multilayer Perceptrons**

### **Definition**

* Depth of Deep Learning refers to the multitude of hidden layers inside Neural Networks;


* Below an illustration of deep Neural network with $5$ hidden layers mapping between $\bf{x} \in {\rm I\!R}^4$ and $\bf{y} \in {\rm I\!R}^3$;

\\

![](https://drive.google.com/uc?export=view&id=1wE0AcdQ6vPMxC4N2pe3OAoVj0142fIzP)


* Neural Networks come with different **architectures** according to the task they have to perform 

\\

* The basic and traditional architecture is the **Multilayer Perceptrons** (also called **Feed Forward**) and corresponds to Neural Neyworks

  * Organized into sequence of **Fully Connected** layers  
  
  * In Fully Connected layers each neuron of layer $l$ is connected to **all** neurons of layer $l-1$;

\\

* MLPs are typically used in problems where:
  * inputs and outputs both have fixed length number of dimensions;
  * inputs $\bf{x}$ are made up with features of heterogeneous nature;

\\

* MLPs solve Regression and Classification problems.





### **Prediction**


* Prediction (or inferring) in MLP is done sequentially as illustrated in the MLP below

\\

* The Neural Network below models regression by mapping $\bf{x} \in {\rm I\!R}^3$ to $y \in {\rm I\!R}$ through one hidden layer $\bf{h}^{[1]} \in {\rm I\!R}^4 $ activated with sigmoid function $\sigma(x)$;



![](https://drive.google.com/uc?export=view&id=18fHwAqVGXmAD8jKZokHu9w7SDiLV1yAa)


* Parameters of this MLP are:

  * $\bf{W}^{[1]} \in {\rm I\!R}^4 \times {\rm I\!R}^3$:

  * $\bf{b}^{[1]} \in {\rm I\!R}^4$: 

  * Hidden representation is obtained with 
  
  $$
  \bf{h}^{[1]} = \sigma(\bf{W}^{[1]} \bf{x} + \bf{b}^{[1]}).
  $$.

  * $\bf{W}^{[2]} \in {\rm I\!R} \times {\rm I\!R}^4$:
    * Note that no activation function is required for $\bf{y}$ because  the model performs regression.
    * In classification setting (logistic or softmax) we must apply  sigmoid or softmax function over neurons of the output layer.  
    
  * $\bf{b}^{[2]}  \in {\rm I\!R}$:
    

\\


* The MLP can compactly be expressed as


$$
\hat{\bf{y}} = \bf{W}^{[2]}.\sigma \left(\bf{W}^{[1]}. \bf{x} + \bf{b}^{[1]} \right) + \bf{b}^{[2]},
$$


## **Training**

### **Backpropagation**

* MLPs and other types of Neural Networks are trained with Gradient-Based Algorithm called **Backpropagation** (popularized by G. Hinton in 1986);


* Backpropagation is Gradient Descent where all ***expressions*** of partial derivatives are obtained **automatically** and **efficiently**;


\\

* In order to illustrate Backpropagation consider the regression MLP defined in the previous section
* Let $\theta = \{\bf{W}^{[1]}, \bf{b}^{[1]}, \bf{W}^{[2]}, \bf{b}^{[2]}\}$ be the set of all parameters. 



\\

* Let $\bf{X}$ and $\bf{y}$ be the training dataset where
  
  *  $\bf{X}$ is $n \times d$ matrix of $n$ input examples described with $d$ features;
  
  *  $\bf{y}$ is vector of outputs associated to $n$ rows of $\bf{X}$.
 

* $\bf{\theta}$ are learned by minimizing the mean squared error Loss Function


\begin{equation}
J(\bf{\theta}) = \frac{1}{n}\sum_{i = 1}^{n}\left(y_i -\hat{y}_i\right)^2.
\end{equation}


* MSE Loss with substituted $\hat{y}_i$  is  

\\

$$
J(\bf{\theta}) =  \frac{1}{n}\sum_{i = 1}^{n}\left(y_i - \bf{W}^{[2]}.\sigma \left(\bf{W}^{[1]}. \bf{x}_{[i,:]} + \bf{b}^{[1]} \right) + \bf{b}^{[2]}\right)^2.
$$


\\


* Just like Gradient Descent, BackPropagation minimizes  $J$  through gradient steps;

\\


* $\nabla_{\bf{\theta}} J$ is the vector with partial derivatives of $J$ w.r.t to **every** parameter in $\bf{\theta}$.

\\

![](https://drive.google.com/uc?export=view&id=1jBdIf7kLTA8TkPf6YMfWqwJLUWqXqBg6)






### **Variants of Backpropagation**

* Deep Learning problems typically learn from important datasets of unstructured data (images, videos, audios, documents);

\\

* Loading the whole dataset into memory can easily become impossible;

\\

* To cope with that limit: 
  * Training Datasets are split into **batches** of fixed size (typically $32, 64, 128,\ldots)$;
  * Each iteration in Batch Gradient Descent relies on different batch to compute $\nabla_{\theta} J$;
  * Traditional **iteration** becomes **epoch** which consists in using every batch once.

\\

* Batch Gradient Descent also comes with variants modifying:
  * Initilization
  * learning rate dynamically
* They all aim for faster convergence as well as avoiding saddle points (local optimum)


* Popular Batch Gradient Descents algorithms are: SGD, Adam and RMSProp.


* For detailed overview and comparison check: https://ruder.io/optimizing-gradient-descent/



## **Neural Networks in Tensorflow/Keras**

### **Toy Example: Neural Network for Regression Problem**

#### **Data Generation**

```python
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

n = 100000
X = np.deg2rad(np.linspace(0,720,n))
y = 1.5*np.sin(X) + (0.45*X - 1)

fig, axes = plt.subplots(figsize = (25,5))
plot = sns.lineplot(X, y)
plot.set_xlabel('X')
plot.set_ylabel('y')

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2)
```

#### **Model Design + Training**

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.optimizers import SGD

input_layer = Input(shape = 1)
h1 = Dense(units = 128, activation = 'relu')(input_layer)
output_layer = Dense(units = 1, activation = 'linear')(h1)

my_nn = Model(input_layer, output_layer)

#Training
my_nn.compile(optimizer = 'rmsprop', loss = 'mse', metrics = ['mse'])
my_nn.fit(X_train, y_train, epochs = 50, batch_size = 32, validation_split = 0.2, verbose = 1)
```

#### **Model Performance**

```python
#Evolution of Optimization through Epochs
fig, axes = plt.subplots(figsize = (20,10))
x_epochs = np.arange(0,50)
sns.lineplot(x_epochs, my_nn.history.history['loss'], color = 'blue', ax = axes, marker = 'o', label = 'loss')
sns.lineplot(x_epochs, my_nn.history.history['val_loss'], color = 'red',ax = axes, marker = 'o',  label = 'val_loss')

axes.set_xlabel('Epoch')
axes.set_ylabel('Loss')
plt.legend(loc = "upper right", prop = {'size' : 25})
```

```python
#Prediction on Test Data
y_hat = my_nn.predict(X_test)
fig, axes = plt.subplots(figsize = (20,8))
sns.lineplot(X_test, y_test, color = 'darkgreen', ax = axes, label = 'Truth')
sns.lineplot(X_test, y_hat.reshape(-1), color = 'red', ax = axes, label = 'Prediction')
axes.set_xlabel('X')
axes.set_ylabel('y')
plt.legend(loc = 'upper left', prop = {'size': 15})
```

### **Neural Network to Classify Fashion Items**

#### **Data Compilation**

```python
import numpy as np
import pandas as pd
from tensorflow.keras import datasets 
import matplotlib.pyplot as plt


def preview_classes(X,y, class_names, nbr_examples_per_class = 10, nbr_classes = 10):
    
    fig, axes = plt.subplots(nbr_classes, nbr_examples_per_class+1, 
                             figsize = (1*nbr_examples_per_class, 1*nbr_classes))
    
    for i in range(nbr_classes):
        y_df = pd.DataFrame({'y':y})
        class_i_random_pick = np.random.choice(y_df[y_df.y == i].index, nbr_examples_per_class, replace = False)
        
        #Class Name
        axes[i, 0].text(0,0.5, class_names[i],  dict(size = 10))
        axes[i, 0].axis('off')

        for j,idx in enumerate(class_i_random_pick):
            img = X[idx,:,:]
            axes[i,j+1].imshow(img, cmap = 'gray')
            axes[i,j+1].axis('off')

#Load it
fashion_mnist = datasets.fashion_mnist
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()
X_train_std, X_test_std = X_train/255, X_test/255

class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

preview_classes(X_train, y_train, class_names, nbr_examples_per_class = 20)
```

#### **Model Design + Training**

```python
from tensorflow.keras.models import Model 
from tensorflow.keras.layers import Input, Flatten, Dense 
from tensorflow.keras.callbacks import EarlyStopping

input_layer = Input(shape = [28,28])
flt_img = Flatten()(input_layer)
h1 = Dense(units = 128, activation = 'relu')(flt_img)
output_layer = Dense(units = 10, activation = 'softmax')(h1)

fashion_item_classifier = Model(input_layer, output_layer) 
fashion_item_classifier.summary()


fashion_item_classifier.compile(loss = 'sparse_categorical_crossentropy', 
                                optimizer = 'rmsprop',
                                metrics = ['accuracy'])

earlystopper = EarlyStopping(monitor = 'val_loss', patience = 3)
fashion_item_classifier.fit(X_train_std, y_train, batch_size = 32, epochs = 30, 
                            validation_split = 0.2,  callbacks = [earlystopper])
```

#### **Model Performance**

```python
#Evolution of Optimization through Epochs
fig, axes = plt.subplots(figsize = (20,10))


loss = fashion_item_classifier.history.history['loss']
val_loss = fashion_item_classifier.history.history['val_loss']
x_epochs = np.arange(0,len(loss))
sns.lineplot(x_epochs, loss , color = 'blue', ax = axes, marker = 'o', label = 'loss')
sns.lineplot(x_epochs, val_loss, color = 'red',ax = axes, marker = 'o',  label = 'val_loss')

axes.set_xlabel('Epoch')
axes.set_ylabel('Loss')
plt.legend(loc = "upper right", prop = {'size' : 25})
```

```python
#keep in mind we have an output layer of 10 neurons
#which correspond to probabilities to each fashion article
#if you want to extract the most likely article select the one with the highest proba 
p_hat = fashion_item_classifier.predict(X_test_std)
y_hat = np.argmax(p_hat, axis = 1)

#to evaluate your classifier run this
test_loss, test_accuracy = fashion_item_classifier.evaluate(X_test_std, y_test)
print('Test Loss = ' + str(test_loss) + ' Test Accuracy = ' + str(test_accuracy))
```
